```python
import librosa
import numpy as np

def get_rms(file_path):
    """Calculates the RMS of a WAV file."""
    y, sr = librosa.load(file_path)  # 필요하면 sr=None으로 바꿔 원본 SR 유지 가능
    rms = np.sqrt(np.mean(y**2))
    return rms

BASE_DIR = "/home/C_Weight/jaewoo_test/00micloaker/records"
file2 = f"/home/C_Weight/jaewoo_test/00micloaker/1002_test/x10_25khz.wav"
#file2 = f"{BASE_DIR}/file48.wav"

# file2는 한 번만 계산
rms2 = get_rms(file2)
print(f"RMS of {file2}: {rms2:.4f}\n")

# file31 ~ file46 순회
for i in range(31, 47):  # 31 to 46 inclusive
    file1 = f"{BASE_DIR}/file{i}.wav"
    rms1 = get_rms(file1)
    sjr_db = 20 * np.log10(rms1 / rms2)
    print(f"RMS of {file1}: {rms1:.4f}")
    print(f"SJR (file{i} vs file47): {sjr_db:.4f} dB\n")
```
